#####**Trends for the Google Trends Dataset**

We gained important insights from EDA about the time series data, such as the observation that South Africa saw a rise in tomato searches over time, which in turn led to an increase in tomato demand from consumers.


```python
#Convert the month variable to the datetime datatype.
df1['Month'] = pd.to_datetime(df1['Month'])
```

```python
#scatterplot
fig, ax = plt.subplots(figsize=(25,6))
ax.scatter(df1['Month'], df1['Tomato Interest'])
ax.set_xlabel('Month')
ax.set_ylabel('Tomato: (South Africa)')
ax.set_title("Trends in South African Tomato Consuption/interest - 2004 to 2023")
plt.show()
```

```python
def plot_df(df1, x, y, title = "", xlabel='Month', ylabel = 'Consumer interest', dpi = 100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color = 'tab:blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
```

```python
plot_df(df1, x=df1['Month'],y=df1['Tomato Interest'],title='Trends in South African Tomato Consuption/interest - 2004 to 2023')
```
